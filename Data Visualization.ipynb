```python
df=pd.read_csv(r'C:\Users\Marys\Downloads\Pokemon1.csv',index_col=0,header=0)
```

```python
#scatter plot is used to visualize the relationship btwn 2 continious variables.

plt.scatter(data=df,x='Attack',y='Defense')
```

```python
plt.figure(figsize=(20,10))
plt.scatter(data=df,x='Attack',y='Defense',s=500,c='purple',marker='*')

#change axes range
plt.xlim(0,150)
plt.ylim(0,200)
#add title
plt.title('Relationship betwen Attack and Defense', size=20)
#add x n y labels
plt.xlabel('Attack Points',size=15)
plt.ylabel('Defense Points',size=15)
plt.grid(True)
#show plot
plt.savefig('test.jpg')
plt.show()

```

```python
#scatter plot with with regression line

import seaborn as sns
sns.lmplot(data=df,x='Attack',y='Defense')
```

```python
sns.lmplot(data=df,x='Attack',y='Defense',fit_reg=False)
```

```python
sns.lmplot(data=df,x='Attack',y='Defense',fit_reg=False,hue='Stage',palette='Set1',markers=['*','^','o'])
```

```python
sns.lmplot(data=df,x='Attack',y='Defense',fit_reg=False,hue='Legendary',palette='Set1',markers=['*','^'])
```

```python
# pair plot when we need more than 1 graph for single output in a perticular df we use pairplot
sns.pairplot(data=df,x_vars=['Attack','Defense','Speed'],
             y_vars=['Total', 'HP'],
             kind='scatter')
```

```python
#Histograms are usefu to visulize the frequency distribution of single continuous variable.
import numpy as np
plt.hist(x='Attack',data=df,bins=10,color="darkblue")
plt.title('Attacks')
plt.show()
```

```python
plt.figure(figsize=(40,10))
plt.subplot(131)   ## 1 row 3clumn 1st position
plt.hist(x="Attack",data=df,bins=10,color='darkblue')
plt.title('Attacks')

plt.subplot(132) ##1 row 3column 2nd position
plt.hist(x="Defense",data=df,bins=10,color='purple')
plt.title('Defense')
plt.xticks(np.arange(0,201,10),rotation=90) #array from 0 to 200 with an increment of 10
plt.yticks(np.arange(0,41,5))

plt.subplot(133) ##1 row 3column 3nd position
plt.hist(x="Speed",data=df,bins=10,color='green')
plt.title('Speed')

plt.show()
           
           
           
           
```

```python
#countplot are useful to visualize the frequency distribution of a simple/discrete/catagorilacl variables

sns.countplot(x='Type 1',data=df,palette='Set3')

#rotate x Labels
plt.xticks(rotation=45)
```

```python
sns.countplot(x='Type 1',data=df,palette='Set3',order=df['Type 1'].value_counts().sort_values(ascending=True).index)
plt.xticks(rotation =45)
```

```python
sns.countplot(x='Type 1',data=df,palette='Set1',order=df['Type 1'].value_counts().index.sort_values(ascending=True))
plt.xticks(rotation =45)
```

```python
#Bar graph is used to visualized  the relationship between 1 catgarical n 1 nymerical variable

sns.barplot(x='Type 1', y = 'Attack',data=df)
plt.xticks(rotation =45)
```

```python
#boxplot is use to visualize the distribution of continues variable but more wisely used to detect the outliers in the variables
#generate boxplots only for selected or individual variables for better reprenstation
sns.boxplot(data=df)
```

```python
#line plot is useful to visulize a continous variable over a period of time eg ecg, stock market prize 
x=[10,30,20,70]
y=[10,45,20,70]
plt.plot(x,y,c='red')

plt.xlim(0,100)
plt.ylim(0,100)
plt.title('A Simple Line Plot')
plt.grid(True)
plt.savefig('test.jpg')
for a,b in zip(x,y):
    plt.text(a,b,(a,b))
plt.show()
```

```python
#Pie chart are useful in visualizing the % distribution of a single catagorical variable
labels = ['SQL','Python','R','Tableau']
sizes = [75,156,40,38]
explode = [0.2,0,0.1,0]
plt.pie(sizes,explode=explode, labels=labels, autopct="%1.2f%%",
        colors=["red","green","blue","purple"],
        shadow= True, startangle=90,counterclock=True)
```
